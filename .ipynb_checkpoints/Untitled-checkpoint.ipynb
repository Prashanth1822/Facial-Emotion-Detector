```python
import cv2
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from keras.models import model_from_json
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img
from tqdm import tqdm

# ULTRA ENHANCED image preprocessing function
def super_enhance_image(img):
    """
    Complete professional enhancement pipeline for crystal clear images
    """
    # Convert to grayscale if needed
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 1. Resize image
    img = cv2.resize(img, (48, 48))
    
    # 2. Noise reduction while preserving edges
    img = cv2.bilateralFilter(img, 9, 75, 75)
    
    # 3. AGGRESSIVE contrast enhancement with CLAHE
    clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(8,8))
    img = clahe.apply(img)
    
    # 4. Global histogram equalization
    img = cv2.equalizeHist(img)
    
    # 5. STRONG sharpening filter
    kernel = np.array([[-1, -1, -1],
                       [-1, 12, -1],
                       [-1, -1, -1]])
    img = cv2.filter2D(img, -1, kernel)
    
    return img

# Enhanced feature extraction function 1
def extract_features(images):
    features = []
    for image in tqdm(images):
        img = load_img(image, grayscale=True)
        img = np.array(img)
        
        # Apply SUPER enhancement
        img_enhanced = super_enhance_image(img)
        
        features.append(img_enhanced)
    
    features = np.array(features)
    features = features.reshape(len(features), 48, 48, 1)
    return features / 255.0

# Enhanced feature extraction function 2
def ef(image):
    img = load_img(image, grayscale=True)
    feature = np.array(img)
    
    # Apply SUPER enhancement
    feature_enhanced = super_enhance_image(feature)
    
    feature_enhanced = feature_enhanced.reshape(1, 48, 48, 1)
    return feature_enhanced / 255.0

# --- Training Data Loading ---
train_root = 'train/'
x_train = []
y_train = []

print("ğŸ”„ Loading and ULTRA-ENHANCING training data...")

for emotion in os.listdir(train_root):
    emotion_folder = os.path.join(train_root, emotion)
    if not os.path.isdir(emotion_folder):
        continue

    image_paths = [os.path.join(emotion_folder, f) for f in os.listdir(emotion_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

    print(f"ğŸ“ Processing train folder: {emotion} ({len(image_paths)} images)")

    for image_path in image_paths:
        img = cv2.imread(image_path)
        if img is None:
            print(f"âš ï¸ Could not read image: {image_path}")
            continue
        
        # Apply SUPER enhancement pipeline
        img_enhanced = super_enhance_image(img)
        
        # Normalize and reshape
        img_enhanced = img_enhanced / 255.0
        img_enhanced = np.reshape(img_enhanced, (1, 48, 48, 1))
        
        x_train.append(img_enhanced)
        y_train.append(emotion)

x_train = np.vstack(x_train)
y_train = np.array(y_train)

# --- Testing Data Loading ---
test_root = 'test/'
x_test = []
y_test = []

print("ğŸ”„ Loading and ULTRA-ENHANCING testing data...")

for emotion in os.listdir(test_root):
    emotion_folder = os.path.join(test_root, emotion)
    if not os.path.isdir(emotion_folder):
        continue

    image_paths = [os.path.join(emotion_folder, f) for f in os.listdir(emotion_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

    print(f"ğŸ“ Processing test folder: {emotion} ({len(image_paths)} images)")

    for image_path in image_paths:
        img = cv2.imread(image_path)
        if img is None:
            print(f"âš ï¸ Could not read image: {image_path}")
            continue
        
        # Apply SUPER enhancement pipeline
        img_enhanced = super_enhance_image(img)
        
        # Normalize and reshape
        img_enhanced = img_enhanced / 255.0
        img_enhanced = np.reshape(img_enhanced, (1, 48, 48, 1))
        
        x_test.append(img_enhanced)
        y_test.append(emotion)

x_test = np.vstack(x_test)
y_test = np.array(y_test)

print("âœ… All data ULTRA-ENHANCED and ready!")
print("x_train shape:", x_train.shape, "y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape, "y_test shape:", y_test.shape)
```

```python
# Label encoding
le = LabelEncoder()
le.fit(y_train)  # Changed from train['label'] to y_train

y_train_encoded = le.transform(y_train)
y_test_encoded = le.transform(y_test)

# Convert to categorical
y_train = to_categorical(y_train_encoded, num_classes=7)
y_test = to_categorical(y_test_encoded, num_classes=7)

# Normalization (if not already done during loading)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Emotion labels
label = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization
from keras.optimizers import Adam
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
import numpy as np

# Fixed model architecture
model = Sequential()

# First convolutional block - START SMALLER
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(48,48,1)))
model.add(BatchNormalization())  # âœ… ADDED: Stabilizes training
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))  # âœ… REDUCED: Less dropout at start

# Second convolutional block - GRADUALLY INCREASE
model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(BatchNormalization())  # âœ… ADDED
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.3))

# Third convolutional block
model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
model.add(BatchNormalization())  # âœ… ADDED
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.3))

# Fourth convolutional block
model.add(Conv2D(256, kernel_size=(3,3), activation='relu'))
model.add(BatchNormalization())  # âœ… ADDED
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Flatten())

# Fully connected layers - SIMPLIFIED
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())  # âœ… ADDED
model.add(Dropout(0.5))

model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())  # âœ… ADDED
model.add(Dropout(0.5))

# Output layer
model.add(Dense(7, activation='softmax'))

# âœ… FIXED COMPILATION
model.compile(
    optimizer=Adam(learning_rate=0.001),  # âœ… SPECIFIC learning rate
    loss='categorical_crossentropy', 
    metrics=['accuracy']
)

model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("ğŸ” CRITICAL DIAGNOSIS - RUN THIS NOW:")

# 1. Check your data shapes and types
print("1. DATA SHAPES:")
print(f"x_train: {x_train.shape}, y_train: {y_train.shape}")
print(f"x_test: {x_test.shape}, y_test: {y_test.shape}")

# 2. Check data ranges
print("\n2. DATA RANGES:")
print(f"x_train: {x_train.min():.3f} to {x_train.max():.3f}")
print(f"x_test: {x_test.min():.3f} to {x_test.max():.3f}")

# 3. Check label distribution
print("\n3. LABEL DISTRIBUTION:")
train_labels = np.argmax(y_train, axis=1) if len(y_train.shape) > 1 else y_train
test_labels = np.argmax(y_test, axis=1) if len(y_test.shape) > 1 else y_test

print("Training labels:", np.unique(train_labels, return_counts=True))
print("Test labels:", np.unique(test_labels, return_counts=True))

# 4. Check if data is normalized
if x_train.max() > 1.0:
    print("âŒ DATA NOT NORMALIZED! Normalizing now...")
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

# 5. Visualize some samples
print("\n4. SAMPLE VISUALIZATION:")
plt.figure(figsize=(12, 4))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(x_train[i].squeeze(), cmap='gray')
    plt.title(f'Label: {train_labels[i]}')
    plt.axis('off')
plt.show()
```
